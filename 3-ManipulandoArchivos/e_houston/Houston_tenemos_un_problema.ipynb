##Asalto Agravado

```python
A = D[D['Offense Type']=='Aggravated Assault']
#Extraer en DataFrame 'A' solo las observaciones de 'Aggravated Assault'
B = A['# Of Offenses'].groupby(A['Hour'])
print(B.sum(),B.mean())
```

Estos resultados indican que en el periodo de observacion, a las 6 de la mañana se presentó el menor numero de Asaltos agravados. Tambien se obtiene que para las observaciones realizadas, en promedio se registra unicamente 1.0 ofensas de Asaltos aagravados a las 6 de la mañana.

En conclusion, la hora mas segura para salir, segun el delito de asalto agravado, es a las **6 de la mañana**
